```python
import os
path = r'bigquery-demo-335405-d7f1212692d0.json'
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = path



import time
from google.cloud import pubsub_v1
project = "bigquery-demo-335405"
pubsub_topic = "projects/bigquery-demo-335405/subscriptions/data_from_file_sub"
input_file = "sales_record.csv"
publisher = pubsub_v1.PublisherClient()
with open(input_file, 'rb') as ifp:
        header = ifp.readline()
        for line in ifp:
            event_data = line
            print('Publishing {0} to {1}'.format(event_data, pubsub_topic))
            publisher.publish(pubsub_topic, event_data)
            time.sleep(1)
```
